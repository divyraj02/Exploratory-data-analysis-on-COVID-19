```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\Lenovo\Downloads\covid_19_india.csv")
df
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python

# Drop missing values or fill them
df = df.dropna()  # or df.fillna(value)

# Remove duplicate rows
df = df.drop_duplicates()

# Convert 'Date' to datetime format and sort by date
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date')
df
```

```python
import numpy as np

# Compute basic statistics for Confirmed, Cured, and Deaths
mean_confirmed = np.mean(df['Confirmed'])
median_confirmed = np.median(df['Confirmed'])
std_dev_confirmed = np.std(df['Confirmed'])

mean_cured = np.mean(df['Cured'])
median_cured = np.median(df['Cured'])
std_dev_cured = np.std(df['Cured'])

mean_deaths = np.mean(df['Deaths'])
median_deaths = np.median(df['Deaths'])
std_dev_deaths = np.std(df['Deaths'])

print(f"Mean confirmed cases: {mean_confirmed}")
print(f"Median confirmed cases: {median_confirmed}")
print(f"Standard deviation of confirmed cases: {std_dev_confirmed}")

print(f"Mean cured cases: {mean_cured}")
print(f"Median cured cases: {median_cured}")
print(f"Standard deviation of cured cases: {std_dev_cured}")

print(f"Mean deaths: {mean_deaths}")
print(f"Median deaths: {median_deaths}")
print(f"Standard deviation of deaths: {std_dev_deaths}")
```

```python
import matplotlib.pyplot as plt

# Plot the number of confirmed cases over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Confirmed'], marker='o', linestyle='-', color='blue', label='Confirmed Cases')
plt.title('COVID-19 Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()

```

```python
# Plot the number of cured cases over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Cured'], marker='o', linestyle='-', color='green', label='Cured Cases')
plt.title('COVID-19 Cured Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cured Cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()

```

```python
# Plot the number of deaths over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Deaths'], marker='o', linestyle='-', color='red', label='Deaths')
plt.title('COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Plot the distribution of confirmed cases
plt.figure(figsize=(8, 6))
plt.hist(df['Confirmed'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of COVID-19 Confirmed Cases')
plt.xlabel('Number of Confirmed Cases')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = df[['Confirmed', 'Cured', 'Deaths']].corr()

# Plot correlation heatmap
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of COVID-19 Cases')
plt.show()
```

```python
state_summary = df.groupby('State/UnionTerritory').agg({
    'Confirmed': 'sum',
    'Cured': 'sum',
    'Deaths': 'sum'
}).reset_index()
```

```python
# Find the top 10 states with the highest number of confirmed cases
top_states = state_summary.nlargest(10, 'Confirmed')

```

```python
# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.bar(top_states['State/UnionTerritory'], top_states['Confirmed'], color='teal')
plt.title('Top 10 States/UTs by Confirmed COVID-19 Cases')
plt.xlabel('State/Union Territory')
plt.ylabel('Number of Confirmed Cases')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
